```python
def fom_plot_setup(font_size=32, label_size=32):
    plt.rc('text', usetex=True)
    plt.rc('font', family='serif')
    ax = plt.gca()
    plt.rc('font', size=font_size)
    ax.yaxis.get_major_formatter().set_powerlimits((0, 1))
    #ax.spines["top"].set_visible(False)  
    #ax.spines["right"].set_visible(False)
    #ax.get_xaxis().tick_bottom()  
    #ax.get_yaxis().tick_left()
    plt.ylabel('Figure of merit', fontsize=label_size)
    plt.xlabel('$n$', fontsize=label_size)
    plt.xticks(fontsize=font_size)
    plt.yticks(fontsize=font_size)
    ax.tick_params(axis='y', labelsize=font_size)
    line = plt.gca().get_lines()[0]
    line.set_marker('.') 
    line.set_color('black')
    plt.grid(True,which='both',color='0.5')
    plt.xscale('linear')
    return plt.gcf()
```

# RATIO PLOTS

Ingest the all the data for the PWR test case

```python
t_wdt = np.linspace(0.1,1.0,10)

base = data_dir + 'pwr/S0100/'
dirs = [base + 'W' + str('{:0>4d}'.format(int(n*1000))) for n in t_wdt]
name = [str(n) for n in t_wdt]
pwr_comp = fom.Comparator(dirs,name)
```

Get FOM data for each data set. $x$ will get the name of the set (the value of $t_{\text{wdt}}$) and $y$ gets the final FOM value (returned by `get_data`, which needs to be sorted)

```python
label = 'INF_FLX'
grp = 1

x = []
y = []
yerr = []

for data_set in pwr_comp.data:
    # Get FOM data and sort by cycle
    data = data_set.get_data(label,grp)
    data = data[data[:,0].argsort()]
    
    x.append(float(data_set.name))
    y.append(data[-1,1])
    
    # Get STD from last half of dataset, square root is required to
    # make it the standard deviation
    yerr.append(np.sqrt(data_set.get_var(label,grp)))
```

Plot the FOM

```python
plt.figure(figsize=(12,9))
plt.errorbar(x,y,yerr=yerr, fmt='k.',ms=12)

fom_plot_setup(12,12)
plt.show()
```

Calculate the ratio of each point to the base case, $t_{\text{wdt}} = 0.1$ which corresponds to no WDT. Propegate the errors of the ratio using:
\begin{align}
r &= \frac{FOM}{FOM_0} \\
\delta r &= r\times \sqrt{\left(\frac{\delta FOM}{FOM}\right)^2 + \left(\frac{\delta FOM_0}{FOM_0}\right)^2}
\end{align}


We can verify that the index `0` corresponds to the base case by checking the $x$ vector

```python
x[0]
```

```python
r = np.ones_like(x)
rerr = np.zeros_like(x)
for i in range(1,len(x)):
    r[i] = y[i]/y[0]
    rerr[i] = r[i]*np.sqrt(np.power(yerr[i]/y[i],2) 
                           + np.power(yerr[0]/y[0],2))
```

```python
fontsize=20
plt.figure(figsize=(12,9))
plt.errorbar(x,r,yerr=rerr, fmt='k.', ms=12, capsize=0)
plt.xlim([0.15,1.05])
fom_plot_setup(fontsize,fontsize)
plt.ylabel('Figure of merit ratio')
plt.xlabel('$t_{\mathrm{wdt}}$', fontsize=fontsize+4)
plt.axhline(y=1.0, ls='--', c='k')
plt.show()
```

Codify those steps in functions:

`get_fom` will return the values of $t_{\mathrm{wdt}}$, FOM and FOM error:

```python
def get_fom(comparator, label, grp):
    x = []
    y = []
    yerr = []
    
    for data_set in comparator.data:
        # Get FOM data and sort by cycle
        data = data_set.get_data(label,grp)
        data = data[data[:,0].argsort()]

        x.append(float(data_set.name))
        y.append(data[-1,1])

        # Get STD from last half of dataset, square root is required to
        # make it the standard deviation
        yerr.append(np.sqrt(data_set.get_var(label,grp)))
        
    return x, y, yerr
```

`get_ratios` will return the values of $t_{\mathrm{wdt}}$, FOM ratios and FOM ratio error:

```python
def get_ratios(comparator, label, grp):
    x, y, yerr = get_fom(comparator, label, grp)
    
    # Find base case
    n = x.index(0.1)
    
    r = np.ones_like(x)
    rerr = np.zeros_like(x)
    
    for i in range(0,len(x)):
        if i != n:
            r[i] = y[i]/y[n]
            rerr[i] = r[i]*np.sqrt(np.power(yerr[i]/y[i],2) 
                                   + np.power(yerr[n]/y[n],2))
            
    return x, r, rerr
```

These will plot the values from the previous functions and save if desired

```python
def plot_fom(comparator, casename, label, grp, save=False, fontsize=20):
   
    if label=='INF_FLX':
        param = ' infinite flux '
    elif label =='INF_TOT':
        param = ' infinite $\Sigma_t$ '
        
    title = casename + param + 'for group ' + str(grp)
    
    x, y, yerr = get_fom(comparator, label, grp)
    
    plt.figure(figsize=(12,9))
    plt.errorbar(x,y,yerr=yerr, fmt='k.',ms=12)
    plt.title(title)
    fom_plot_setup(fontsize,fontsize)
    plt.xlim([0.05,1.05])
    plt.xticks(np.arange(0.1,1.1, 0.1))
    plt.show()
```

```python
def plot_ratios(comparator, casename, label, grp, save=False, fontsize=20):
    if label=='INF_FLX':
        param = ' infinite flux '
    elif label =='INF_TOT':
        param = ' infinite $\Sigma_t$ '
        
    title = casename + param + 'for group ' + str(grp)
    filename = casename.lower() + '_' + label.lower() + '_' + str(grp)

    x, r, rerr = get_ratios(comparator, label, grp)

    plt.figure(figsize=(12,9))
    plt.errorbar(x,r,yerr=rerr, fmt='k.', ms=12, capsize=0)
    fom_plot_setup(fontsize,fontsize)
    plt.xticks(np.arange(0.1,1.1, 0.1))
    plt.xlim([0.15,1.05])
    plt.title(title)
    plt.ylabel('Figure of merit ratio')
    plt.xlabel('$t_{\mathrm{wdt}}$', fontsize=fontsize+4)
    plt.axhline(y=1.0, ls='--', c='k')
    if save:
        plt.savefig(img_dir + filename + ".pdf", 
                    format = 'pdf', bbox_inches='tight')
    else:
        plt.show()
```

```python
def pandas_table(x, y, yerr, r, rerr):
    d = {'twdt' : x, 'fom' : y, 'fom_err': yerr, 'r' : r, 'r_err' : rerr}
    df = pd.DataFrame(d)
    #Move twdt to the front
    cols = df.columns.tolist()
    cols = cols[-1:] + cols[:-1]
    df = df[cols]
    return df
```

```python
def pandas_format(df, fom_p, rat_p):
    # Formats the data frame
    def fom_form(x):
        return '{:.0f}'.format(np.around(x/np.power(10,fom_p)))
    def ratio(x):
        return '{:.3f}'.format(x)
    df['fom'] = df['fom'].apply(fom_form)
    df['fom_err'] = df['fom_err'].apply(fom_form)
    df['r']  = df['r'].apply(ratio)
    df['r_err']  = df['r_err'].apply(ratio)
```

```python
def latex(df, fom_p, rat_p):
    pandas_format(df, fom_p, rat_p)
    df.insert(2,'fom_pm',['$\pm$']*len(x))
    df.insert(5,'r_pm',['$\pm$']*len(x))
    def f(x): 
        return x
    print df.to_latex(index=False, escape=False, column_format='rrcrrcr')
```

# CYCLES/CPU

Loading on the machine may cause the simulation to slow down, affecting the FOM. It is therefore important to verify that cycles require a consistent amount of CPU time, as much as possible.

We need to access the `DataFile` within each `Analyzer` that is in the `Comparater.` The Analyzer has all the data from one value of $t_{\mathrm{wdt}}$, so we need to verify we have the right index.

```python
pwr_comp.data[0].name
```

Here we are looking at the base case. We will now cycle through all the `DataFile` objects and get the value of the cycles (`CYCLE_IDX`) and the CPU time.

```python
cycles = []
cpu = []
for data in pwr_comp.data[0].data:
    cycles.append(data.get_data('CYCLE_IDX')[0][0])
    cpu.append(data.get_cpu())
    
cyc_cpu = np.column_stack((cycles, cpu))
    
cyc_cpu = cyc_cpu[cyc_cpu[:,0].argsort()]
```

What we are interested in is the differential values:
\begin{equation}
\frac{\mathrm{cycles}_{i} - \mathrm{cycles}_{i-1}}{\mathrm{CPU}_i - \mathrm{CPU}_{i-1}}
\end{equation}

```python
dcyc_cpu = []

for i in range(1,len(cycles)):
    dcyc = cyc_cpu[i,0] - cyc_cpu[i-1,0]
    dcpu = cyc_cpu[i,1] - cyc_cpu[i-1,1]
    dcyc_cpu.append(dcyc/dcpu)
```

```python
plt.plot(dcyc_cpu, 'k.')
fom_plot_setup(12,12)
plt.show()
```

All contained in one function:

```python
def cyc_cpu_plot(comp, twdt):
    # Find index
    for i, analyzer in enumerate(comp.data):
        if analyzer.name == str(twdt):
            n = i
    cycles = []
    cpu = []
    for data in comp.data[n].data:
        cycles.append(data.get_data('CYCLE_IDX')[0][0])
        cpu.append(data.get_cpu())
    
    cyc_cpu = np.column_stack((cycles, cpu))
    
    cyc_cpu = cyc_cpu[cyc_cpu[:,0].argsort()]
    dcyc_cpu = []

    for i in range(1,len(cycles)):
        dcyc = cyc_cpu[i,0] - cyc_cpu[i-1,0]
        dcpu = cyc_cpu[i,1] - cyc_cpu[i-1,1]
        dcyc_cpu.append(dcyc/dcpu)
    plt.title('Cycles/CPU time for ' + comp.data[n].name)
    plt.plot(dcyc_cpu, 'k.')
    fom_plot_setup(12,12)
    plt.ylabel('Cycles/CPU time')
    plt.show()
                        
```

```python
def get_cyc_cpu(comp, twdt):
    for i, analyzer in enumerate(comp.data):
        if analyzer.name == str(twdt):
            n = i
    cycles = []
    cpu = []
    for data in comp.data[n].data:
        cycles.append(data.get_data('CYCLE_IDX')[0][0])
        cpu.append(data.get_cpu())
    
    cyc_cpu = np.column_stack((cycles, cpu))
    
    cyc_cpu = cyc_cpu[cyc_cpu[:,0].argsort()]
    dcyc_cpu = []

    for i in range(1,len(cycles)):
        cycs = np.sum(cyc_cpu[0:i,0])
        cpus = np.sum(cyc_cpu[0:i,0])
        dcyc_cpu.append(cycs/cpus)
    #    dcyc = cyc_cpu[i,0] - cyc_cpu[i-1,0]
    #    dcpu = cyc_cpu[i,1] - cyc_cpu[i-1,1]
    #    dcyc_cpu.append(dcyc/dcpu)
        
    return dcyc_cpu
```

```python
def get_corr_fom(comp, label, grp, twdt):
    for i, analyzer in enumerate(comp.data):
        if analyzer.name == str(twdt):
            n = i
    data = comp.data[n].get_data(label,grp,fom=False)
    cyc_cpu = get_cyc_cpu(comp,twdt)
    
    assert np.shape(data)[0] == np.shape(cyc_cpu)[0] + 1, "something went wrong with the array size"
    
    # Calculate FOM
    fom = np.zeros_like(data)
    data = data[data[:,0].argsort()]
    
    fom[:,0] = data[:,0]
    denom = np.multiply(np.power(data[:,1],2),data[:,0])
    fom[1:,1] = np.multiply(np.power(denom[1:], -1),cyc_cpu)
    #fom[1:,1] = np.multiply(np.power(denom[1:], -1),np.average(cyc_cpu))
    
    yerr = np.sqrt(np.var(fom[int(np.ceil(np.shape(fom)[0]/2.0)):,1])) 
    
    return fom[1:,:], yerr
```
